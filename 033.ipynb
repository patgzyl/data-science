```
df=fetch_finacial_data('UBER')
df.info()
```

```
df['Avr']=(df.Open+df.Close)/2.0
df.head()
```

```
df.sort_index()
```

```
df['Close_shitf']=df.Close.shift(1)
df.head()
```

```
df['Daily_change']=df.Close/df.Close_shitf - 1
df.head()
```
